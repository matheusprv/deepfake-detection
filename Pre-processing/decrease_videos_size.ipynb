```python
import os
import shutil
import cv2
```

```python
#For real videos
def split_videos(folder_path, output_folder):

    # Iterate through all files in the folder
    files_list = os.listdir(folder_path)
    total_files = len(files_list)

    for index, filename in enumerate(files_list):
        
        print(f"{index+1}/{total_files}", end=" ")
        
        video_path = os.path.join(folder_path, filename)
        cap = cv2.VideoCapture(video_path)
        
        # Get total number of frames in the video
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        
        # Define the number of segments and frames per segment
        num_segments = 6
        frames_per_segment = 24

        if total_frames < 144:
            num_segments = total_frames // frames_per_segment
        
        # Extract and save segmented videos
        for i in range(num_segments):
            frames = []
            for j in range(frames_per_segment):
                ret, frame = cap.read()
                if ret:
                    frames.append(frame)
            
            # Write frames to a new video file
            output_video_path = os.path.join(output_folder, f"{filename.split('.')[0]}_segment_{i+1}.mp4")
            print(output_video_path)
            out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), 30, (frames[0].shape[1], frames[0].shape[0]))
            for frame in frames:
                out.write(frame)
            out.release()
                
        cap.release()
```
